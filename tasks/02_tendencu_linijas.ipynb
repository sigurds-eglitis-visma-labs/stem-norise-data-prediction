### 2.1 ImportÄ“jam bibliotÄ“kas

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = [12, 6]
````

### 2.2 IelÄdÄ“jam un sagatavojam datus

````python
# IelÄdÄ“jam datus
df = pd.read_csv('../data/coffee_sales/index_1.csv')
df['date'] = pd.to_datetime(df['date'])

# AprÄ“Ä·inÄm dienas ieÅ†Ä“mumus
daily_revenue = df.groupby('date')['money'].sum().reset_index()
daily_revenue.columns = ['date', 'revenue']

# Pievienojam dienu numuru (lai varÄ“tu izmantot regresijÄ)
daily_revenue['day_number'] = range(len(daily_revenue))

daily_revenue.head(10)
````

### 2.3 Kas ir tendenÄu lÄ«nija?

**TendenÄu lÄ«nija** (trend line) parÄda datu vispÄrÄ“jo virzienu.

VienkÄrÅ¡ÄkÄ metode ir **lineÄrÄ regresija** - mÄ“s meklÄ“jam taisni, kas vislabÄk "iederas" mÅ«su datos.

MatemÄtiski: `y = mx + b`
- `m` - slÄ«pums (cik strauji mainÄs)
- `b` - sÄkuma punkts

ğŸ“– [VairÄk par lineÄro regresiju](https://scikit-learn.org/stable/modules/linear_model.html#ordinary-least-squares)

````python
# Izveidojam lineÄrÄs regresijas modeli
X = daily_revenue['day_number'].values.reshape(-1, 1)  # Ieejas dati
y = daily_revenue['revenue'].values  # MÄ“rÄ·a dati

# ApmÄcÄm modeli
model = LinearRegression()
model.fit(X, y)

# ParÄdÄm rezultÄtus
print(f"SlÄ«pums (m): {model.coef_[0]:.4f}")
print(f"SÄkuma punkts (b): {model.intercept_:.4f}")
print(f"\nTas nozÄ«mÄ“: katru dienu ieÅ†Ä“mumi vidÄ“ji mainÄs par {model.coef_[0]:.2f} EUR")
````

### 2.4 ZÄ«mÄ“jam tendenÄu lÄ«niju

````python
# AprÄ“Ä·inÄm tendenÄu lÄ«nijas vÄ“rtÄ«bas
trend_line = model.predict(X)

# ZÄ«mÄ“jam grafiku
plt.figure(figsize=(14, 6))
plt.plot(daily_revenue['date'], daily_revenue['revenue'], 'b-', alpha=0.5, label='Faktiskie ieÅ†Ä“mumi')
plt.plot(daily_revenue['date'], trend_line, 'r-', linewidth=2, label='TendenÄu lÄ«nija')
plt.title('Dienas ieÅ†Ä“mumi ar tendenÄu lÄ«niju')
plt.xlabel('Datums')
plt.ylabel('IeÅ†Ä“mumi (UAH)')
plt.legend()
plt.tight_layout()
plt.show()
````
